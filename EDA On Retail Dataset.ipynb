**Perform Univariate analysis for Categorical features**:

```python
numerical_df = df[numerical_columns]
numerical_df.head()
```

```python

plt.figure(figsize=(15, 20))

# Loop through numerical columns and create boxplots
for i, column in enumerate(numerical_columns):
    plt.subplot(len(numerical_columns), 1, i + 1)
    sns.boxplot(data=df, x=column, color='skyblue')
    plt.title(f'Boxplot of {column}')
    plt.xlabel(column)

# Adjust layout and show the plot
plt.tight_layout()
plt.show()

# Distribution plots
plt.figure(figsize=(15, 20))

for i, column in enumerate(numerical_columns):
    plt.subplot(len(numerical_columns), 1, i + 1)
    sns.histplot(data=df, x=column, kde=True, color='purple', bins=30)
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)

# Adjust layout and show the plot
plt.tight_layout()
plt.show()
```

**Notes**:
- Numerical colums included above are the columns with numerical records such as Listprice, OrderQuantity, UnitPrice, SalesAmount, etc
- The graphical representation above shows the presence of outliers in the numerical columns.

#### Perform time-series analysis:

- Extract insights from Orderdate, Duedate, and Shipdate:
- Find the average shipping time.
- Plot sales trends over time.

```python
# Extract insights from Orderdate, Duedate, and Shipdate:
for column in date_columns:
    unique_values = df[column].nunique()
    missing_values = df[column].isnull().sum()
    print(f"{column}: Unique Values = {unique_values}, Missing Values = {missing_values}")

# 2. Find the average shipping time
# Calculate the shipping time as the difference between Shipdate and Orderdate
df['ShippingTime'] = (df['Shipdate'] - df['Orderdate']).dt.days

# Calculate average shipping time
average_shipping_time = df['ShippingTime'].mean()
print(f"Average Shipping Time: {average_shipping_time:.2f} days")

# Plot sales trends over time
# Aggregate sales by Orderdate
sales_trends = df.groupby('Orderdate')['SalesAmount'].sum().reset_index()

# Plot the sales trends
plt.figure(figsize=(12, 6))
sns.lineplot(x='Orderdate', y='SalesAmount', data=sales_trends, marker='o', color='blue')
plt.title('Sales Trends Over Time')
plt.xlabel('Orderdate', fontsize=14)
plt.ylabel('Sales Amount', fontsize=14)
plt.grid(True)
plt.show()
```
